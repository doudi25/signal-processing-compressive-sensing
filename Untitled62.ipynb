```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from scipy.fftpack import dct,idct
# Generate signal
n = 1000  # points in high resolution signal
t = np.linspace(0, 0.03, n)
f = np.cos(2 * 97 * np.pi * t) + np.cos(2 * 777 * np.pi * t)

# Plot the first 100 points of the signal
plt.plot(t[:100], f[:100], color='red')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Signal composed of two cosine waves')
plt.show()  # Display the plot
```

```python
sampling = 60  # Number of samples to be taken

# Generate random indices for sampling
perm = np.floor(np.random.rand(sampling) * n).astype(int)
# Set the random sampling vector
random_samp=f[perm]
print("Randomly sampled indices:", perm)
```

```python

# Generate the plot
plt.plot(t, f, color='red')  # Plot the original signal
plt.scatter(t[perm],random_samp, color='blue', label='Randomly Sampled Points')  # Scatter plot of randomly sampled points
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Original Signal with Randomly Sampled Points')
plt.legend()  # Show legend
plt.show()
```

```python
# Generate Discrete cosine transform  matrix
psi = dct(np.identity(len(t)))

# Select rows from DCT matrix corresponding to randomly sampled indices
phi = psi[perm, :]

# Initialize Lasso regression model with regularization parameter
model = Lasso(alpha=0.2)

# Fit the Lasso model to the measurements and the randomly sampled signal
model.fit(phi, random_samp)
```

```python
# Obtain sparse representation from Lasso model coefficients
sparse_representation = model.coef_

# Reconstruct the signal using Inverse Discrete Cosine Transform (IDCT)
reconstruction_signal = idct(sparse_representation)
```

```python
# Create subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
# Plot original signal on the first subplot (ax[0])
ax[0].plot(t, f, color='red')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[0].set_title('Original Signal')

# Plot reconstructed signal on the second subplot (ax[1])
ax[1].plot(t, reconstruction_signal, color='blue')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Amplitude')
ax[1].set_title('Reconstructed Signal')

# Display the plot
plt.show()
```

```python

# Calculate mean squared error (MSE)
mse = np.mean((f - reconstruction_signal) ** 2)

# Print the result
print(f'The mean squared error is {mse}')
# You are free to play with the parameters : sampling to be taken and the alpha parameter to get the best fit
```
